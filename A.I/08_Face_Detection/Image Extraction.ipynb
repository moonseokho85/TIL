```
%matplotlib inline
import matplotlib.pyplot as plt
import cv2
import os
```

```
# 캐스케이드 파일 지정해서 검출기 생성하기
cascade_file = "/content/drive/My Drive/Colab Notebooks/Face_Detection/haarcascade_frontface.xml"
cascade = cv2.CascadeClassifier(cascade_file)
```

```
path_daniel = '/content/drive/My Drive/Colab Notebooks/Face_Detection/images/Daniel Craig'
```

```
path_chris = '/content/drive/My Drive/Colab Notebooks/Face_Detection/images/Chris Evans'
```

```
for i, img in enumerate(os.listdir(path_chris)):
  # 이미지를 읽어 들이고 그레이스케일로 변환하기
  img = cv2.imread("/content/drive/My Drive/Colab Notebooks/Face_Detection/images/Chris Evans/"+img)
  try:
    img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # 얼굴 인식하기
    face_list = cascade.detectMultiScale(img_gray)

    # 결과 확인하기
    if len(face_list) == 0:
      print("실패")
      quit()

    # 인식한 부분 표시하기
    for (x,y,w,h) in face_list:  
      print("얼굴의 좌표 =", x, y, w, h)
      # red = (0, 0, 255)
      # cv2.rectangle(img, (x, y), (x+w, y+h), red, thickness=10)
      roi = img[y:y+h, x:x+w]

      # plt.imshow(cv2.cvtColor(roi, cv2.COLOR_BGR2RGB))
      # plt.show()

    cv2.imwrite("/content/drive/My Drive/Colab Notebooks/Face_Detection/images/processed_Chris/Chris Evans"+str(i)+".png", roi)
    print('saved')
    #plt.imshow(cv2.cvtColor(roi, cv2.COLOR_BGR2RGB))
    #plt.show()
  except:
    pass
  
  
```
